Once again we are using the print command.  The operations inside of the () will be performed and then printed.  When there is a comma separating arguements, Python inserts a space.

__Line 1)__  We start with a variable assignment.  We are assigning the string 'spam' to the variable s.  Note that anything enclosed inside of quotation marks (",',''', or """) is considered type string by python.  For example, 5 is a type int while '5' is a type string.

__Line 2)__  A string is an ordered type in Python.  Each character/element of the string is indexed with a unique location that can be accessed by enclosing the index position inside of [ ] brackets following the variable name.  

In python, the first position is 0, followed by 1, then 2, and so on.  This means that for the variable s, the following indexing options are available: s[0] is s (the string character, not the variable), s[1] is p, s[2] is a, s[3] is m.

Python also allows negative indexing where -1 refers to the last character/element of the variable.  Using this method, -2 is the second to last element, -3 is third to last, and so on.  This means that for the variable s, the following index options are available using negative indexing: s[-1] is m, s[-2] is a, s[-3] is p, s[-4] is s (the string character,not the variable). 

When a comma separates two arguements inside of the print command, Python inserts a space.  


__Line 3)__ In line 2 we were indexing into the string s and accessing specific characters.  Python also allows us to index and access ranges of characters/elements.  

The format of an indexing command that contains a range arguement is variable_name[a:b:c] where a=start, b=stop, and c=step

In this construction, variable_name is the name of whatever variable you are working with.  In our case, variable_name is equal to s.  

Working with my previous example, my_variable[a=start,b=stop,c=step] the letter a represents the first character of the range that you are requesting.  If it is left blank, then python defaults to a=0 and the first character will be used.

The b represents the end point of the range but Python does not include the endpoint in the returned answer.  The last element of the range will be the element in the position b-1.  If b is left blank, the last character/element of the string will be included.

The c represents the step value and is assumed to be 1 if left blank. 


s[1:3] is asking for the characters contained in index positions 1 through 2, which is the second and third character.
There are several things to discuss here:

First, the c position is blank.  This means that our indexing operation is using the default step = 1.

Second, notice that I said the characters in index positions 1 through 2 are being returned, not 1 through 3.  As we stated above, 3 is the end point and is not included in the returned product.  Instead, the last character returned is in position 3 - 1 = 2nd position.  

Third, we have to be careful when deciphering which characters are in positions 1 through 2.  Remembering what I said earlier about the index values used in python, 0 represents the first character, 1 represents the second, and 2 represents the third.  Then s[1:3] returns the second and third character.


s[1:] returns the second character through the last character.
The steps for dechipering this are as follows:

First, the c position is blank so the step defaults to 1.

Second, the index value of 1 represents the second character.

Third, the b position is blank so the range goes through and includes the last character.


s[:-1] returns the first through the second to last character in the string.  The steps for dechipering this are as follows:

First, as above, the c position is blank so the step defaults to 1.

Second, the a posistion is blank so the first character is the beginning of the range.

Third, negative indexing is utilized for the b posisiton.  We know that b is the endpoint and does not actually appear in the returned value, instead, the character in index position b-1 is used.  In this case, b-1 =-1-1=-2.  The character which is second from the end has index position -2. 

```python
s = 'spam'
print(s[0], s[-2])
print(s[1:3], s[1:], s[:-1])
```

__Line 3)__ myfile = open('filename','mode') is a command that creates a variable called myfile and assigns it a file.  If no path is specified, the file will be located in the current working directory.  If you specify a path (ie 'C:folder_name/file_name.txt') then Python will open/create a file in that location.  

The first arguement, listed here is the filename that you either want to access of create.  The filename needs to be inside of quotation marks.  The second arguement designates the mode that you are opening the file in.  Common modes are 'r' for reading, 'w' for writing, and 'a' for appending.  

'r'- The reading mode will not change the contents of the file.  You can only read the contents.  If the file does not exist, you will receive an error message.

'w' - The writing mode will create the file if it does not exist.  If the file does exist, the contents of the file will be overwritten with new information.

'a'- The append mode will create the file if it does not exist.  If the file does exist, the contents of the file will not be cleared.  New information will be added to the end of the existing file.

Our command is myfile = open('schedule.txt', 'wb').  This means that we have created a variable called myfile and we have assigned a file named schedule.txt to it.  I could have specified a path but I chose to have this file created in my current working directory.  The mode chosen is 'wb' which is a type of writing mode.  The additional b means that we will be writing in binary.  The binary mode means that we will be writing in bytes without any decoding.

__Line 4)__ Now that we are in the binary writing mode, we need to give Python some information to write to our file.  The command write(r.read()) tells Python to write the entire contents of our variable r into our new variable myfile.  When reading a file, commonly used commands include read(), read(x), and readline().  The read() function that we just used reads the entire contents of the file.  The read(x) function will read the next x characters of the file, where x is a positive integer.  The readline() function reads the entire line as a string.

__Line 5)__  myfile.close() closes the file that we have been using.  Now 'schedule.txt' can be accessed by other programs.

## Show the difference in scope between importing a module and importing a specific variable from the module

In the block of code below, we use a different method of importing a module.  I will contrast two mehtods below:

import module_name:  When we use the import Class_1_Module function convention like we did in the block of code above, all of the attributes of that module must be accessed using the module_name.information_name convention.  As we saw above, the variable title = 'Math 510 Fall 2015' was local to the module and could only be imported into the module by typing Class_1_Module.title.  The value 'Math 510 Fall 2015' was not assigned to any variable in the global space.  


from module_name import information:  This method imports a specific variable from the module into the global scope.  By typing from Class_1_Module import title, we are assign the global variable title the value of the module variable title.  If the global variable title had a value previously, this operation will overwrite its value with the value of the the module variable title.  Once this has been done, 'Math 510 Fall 2015' can be accessed by typing title.  There is no need for the module_name.information_name convention that we have described earlier.  

It is also possible to import all of the information contained inside of a module.  For example, by typing from math import *, we make available all of the varialbes and functions contained within the math module.  Now, typing pi gives us the value of pi as 3.141592. . ., typing e gives us the value of e as 2.718281. . ., typing sqrt(4) give us 2. There is no need to type math.pi, math.e, or math.sqrt().

__Line 3)__  Using the second style described above, from Class_1_Module import title overwrites any global value previously assigned to the variable identifier title and assigns it the value of title within the module.

__Line 4)__ This prints the current value of the variable title which is the value assigned in the previous line.  The previous value of title = 'something else' no longer exists.
